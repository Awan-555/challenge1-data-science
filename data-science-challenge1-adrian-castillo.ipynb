### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]

nombres_tiendas = []
ingresos_totales = []

for i, tienda in enumerate(tiendas, start=1):
    ingresos = tienda['Precio'].sum()
    print(f'Los ingresos totales de la Tienda {i} son de {ingresos}')
    nombres_tiendas.append(f'Tienda {i}')
    ingresos_totales.append(ingresos)

# Gráfico de barras
plt.figure(figsize=(5, 3))
plt.bar(nombres_tiendas, ingresos_totales, color='brown')
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# 2. Ventas por categoría

```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]

for i, tienda in enumerate(tiendas, start=1):
    print(f'Tienda {i}')

    categoria_ventas = tienda[['Categoría del Producto', 'Precio']]\
        .groupby('Categoría del Producto')\
        .sum()

    print(categoria_ventas)
    print()

    total_tienda = categoria_ventas['Precio'].sum()
    print(f'Total en Tienda {i}: {total_tienda}\n')

    # Gráfico de pastel
    plt.figure(figsize=(8, 5))
    categoria_ventas['Precio'].plot(kind='bar')
    plt.grid(axis='y')
    plt.title(f'Ventas por Categoría - Tienda {i}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]

nombres_tiendas = []
promedio_calif = []

for i, tienda in enumerate(tiendas, start=1):
  print(f'Tienda {i}')
  calificacion_promedio_tienda = tienda['Calificación'].mean()
  print(f'Calificación promedio: {calificacion_promedio_tienda:.2f}')
  print()
  nombres_tiendas.append(f'Tienda {i}')
  promedio_calif.append(calificacion_promedio_tienda)

#Grafico de barras
plt.figure(figsize=(4, 4))
plt.bar(nombres_tiendas, promedio_calif, color='lightblue')
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tiendas')
plt.ylim(0, 5)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]

for i, tienda in enumerate(tiendas, start=1):
    print(f'Tienda {i}')
    conteo_productos = tienda['Producto'].value_counts()

    producto_mas_vendido = conteo_productos.idxmax()
    mayor_venta = conteo_productos.max()
    producto_menos_vendido = conteo_productos.idxmin()
    menor_venta = conteo_productos.min()

    print(f"Producto más vendido: {producto_mas_vendido} con {mayor_venta} ventas")
    print(f"Producto menos vendido: {producto_menos_vendido} con {menor_venta} ventas")
    print()

    # Gráfico de barras
    plt.figure()
    plt.bar([producto_mas_vendido, producto_menos_vendido], [mayor_venta, menor_venta], color=['green', 'red'])
    plt.title(f'Tienda {i} - Más y Menos Vendido')
    plt.ylabel('Cantidad de ventas')
    plt.xlabel('Producto')
    plt.tight_layout()
    plt.show()
```

# 5. Envío promedio por tienda

```python
import matplotlib.pyplot as plt

tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = []
promedio_envio = []

for i, tienda in enumerate(tiendas, start=1):
  print(f'Tienda {i}')
  envio_promedio_tienda = tienda['Costo de envío'].mean()
  print(f'Promedio de envío: {envio_promedio_tienda:.2f}')
  print()
  nombres_tiendas.append(f'Tienda {i}')
  promedio_envio.append(envio_promedio_tienda)

#Grafico de barras
plt.figure(figsize=(4, 4))
plt.bar(nombres_tiendas, promedio_envio, color='orange')
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tiendas')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
